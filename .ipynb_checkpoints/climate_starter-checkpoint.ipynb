```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
import datetime as dt
```

```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
Measurement = Base.classes.measurement
Station = Base.classes.station
session = Session(engine)
```

```python
Data = session.query(Measurement).limit(5).all
Data.__dict__
for row in session.query(Measurement).filter(Measurement.date).order_by(Measurement.station).limit(0):
    print(row.__dict__)

data = session.query(Measurement.prcp, Measurement.date).filter(Measurement.date >= "2010-01-01").filter(Measurement.date <= '2010-12-31').all()

max_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
Date = []
Date = list(max_date)
for word in Date:
    year = word[0:4]
    month = word[5:7]
    day = word[8:10]
Max_Date = dt.date(int(year), int(month), int(day))
Previous_Year = Max_Date - dt.timedelta(days=365)
print(f'{Max_Date} {Previous_Year}')

Precipitation_Info = []
Precipitation_Date = []
Precipitation_Data = session.query(Measurement.prcp, Measurement.date).\
    filter(Measurement.date >= Previous_Year).filter(Measurement.date <= Max_Date)

for row in Precipitation_Data:
    Precipitation_Info.append(row[0])
    Precipitation_Date.append(row[1])

df = pd.DataFrame({"Date": Precipitation_Date,
                  "Pecipitation": Precipitation_Info})

df1 = df.dropna(how='any')
df1 = df1.set_index("Date")
df1 = df1.sort_values("Date")

df1.plot(grid=True, rot=100, alpha=10, linewidth=5)
plt.savefig("Images/Climate.png")
plt.show()
```

```python
def calc_temps(start_date, end_date):
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
trip_end_date = '2017-12-31'
year = trip_end_date[0:4]
month = trip_end_date[5:7]
day = trip_end_date[8:10]

trip_end_date = dt.date(int(year), int(month), int(day))
trip_start_date = trip_end_date - dt.timedelta(days=365)

tobs_data = calc_temps(trip_start_date, trip_end_date)
tobs_data = list(np.ravel(tobs_data))
min_tobs = int(tobs_data[0])
avg_tobs = int(tobs_data[1])
max_tobs = int(tobs_data[2])
print(f'{min_tobs} {avg_tobs} {max_tobs}')
```

```python
fig, ax = plt.subplots()
index = 1
bar_width = 0.1

error_config = {'ecolor':'0.2'}

ax.bar(index, avg_tobs, bar_width, color='r', yerr=(max_tobs - min_tobs), error_kw=error_config)
ax.set_ylabel('Temp(F)')
ax.set_title('Trip Avg Temp')
ax.set_xlim(0.8,1.2)

# plt.tight_layout()
plt.savefig("Images/Trip Average Temp.png")
plt.show()
```

```python
def daily_normals(date):    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
